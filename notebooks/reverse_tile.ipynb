```python
import glob
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
ylist = glob.glob("/hdd/Datasets/sar2opt_raw_data/raw_sat_test/baoding/s2/tiles/*")
xlist = glob.glob(f"{ylist[0]}/*.png")
```

```python
x_ext_list = []
for x in xlist:
    x = x.split("/")[-1]
    x_ext_list.append(x)
    
xlist = x_ext_list
```

```python
xlist
```

```python
ylist
```

```python
xlist = sorted(xlist,key=lambda x: int(x.split(".")[0]))
ylist = sorted(ylist,key=lambda y: int(y.split("/")[-1]))
```

```python
retiled_img = None
for y in ylist:
    img_row = None
    for x in xlist:
        img = cv2.imread(f"{y}/{x}")
        cropped_img = img[50:-50, 50:-50]
        if img_row is None:
            img_row = cropped_img
        else:
            img_row = cv2.hconcat([img_row, cropped_img])
    if retiled_img is None:
        retiled_img = img_row
    else:
        retiled_img = cv2.vconcat([retiled_img, img_row])
    
```

```python
plt.figure(figsize=(10,10))
plt.imshow(retiled_img)
```
